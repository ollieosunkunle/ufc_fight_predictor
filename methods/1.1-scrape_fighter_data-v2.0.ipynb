```python
def scrape_fighter_data():
    # Import scrapy
    import scrapy

    # Import the CrawlerProcess
    from scrapy.crawler import CrawlerProcess

    from scrapy import Selector
    import requests

    import pandas as pd
    import numpy as np
    import datetime

    import warnings
    warnings.filterwarnings("ignore")

    page_links = [
        'http://ufcstats.com/statistics/fighters?char=a&page=all',
        'http://ufcstats.com/statistics/fighters?char=b&page=all',
        'http://ufcstats.com/statistics/fighters?char=c&page=all',
        'http://ufcstats.com/statistics/fighters?char=d&page=all',
        'http://ufcstats.com/statistics/fighters?char=e&page=all',
        'http://ufcstats.com/statistics/fighters?char=f&page=all',
        'http://ufcstats.com/statistics/fighters?char=g&page=all',
        'http://ufcstats.com/statistics/fighters?char=h&page=all',
        'http://ufcstats.com/statistics/fighters?char=i&page=all',
        'http://ufcstats.com/statistics/fighters?char=j&page=all',
        'http://ufcstats.com/statistics/fighters?char=k&page=all',
        'http://ufcstats.com/statistics/fighters?char=l&page=all',
        'http://ufcstats.com/statistics/fighters?char=m&page=all',
        'http://ufcstats.com/statistics/fighters?char=n&page=all',
        'http://ufcstats.com/statistics/fighters?char=o&page=all',
        'http://ufcstats.com/statistics/fighters?char=p&page=all',
        'http://ufcstats.com/statistics/fighters?char=q&page=all',
        'http://ufcstats.com/statistics/fighters?char=r&page=all',
        'http://ufcstats.com/statistics/fighters?char=s&page=all',
        'http://ufcstats.com/statistics/fighters?char=t&page=all',
        'http://ufcstats.com/statistics/fighters?char=u&page=all',
        'http://ufcstats.com/statistics/fighters?char=v&page=all',
        'http://ufcstats.com/statistics/fighters?char=w&page=all',
        'http://ufcstats.com/statistics/fighters?char=x&page=all',
        'http://ufcstats.com/statistics/fighters?char=y&page=all',
        'http://ufcstats.com/statistics/fighters?char=z&page=all'

    ]

    # Create the Spider class
    class ufc_fighter_scraper(scrapy.Spider):
        name = 'ufc_fighter_scraper'

        # start_requests method
        def start_requests( self ):
            for url in page_links:
                yield scrapy.Request( url = url, callback = self.parse_front )

        def parse_front(self, response):
            links = response.xpath('/html/body/section/div/div/div/table/tbody/tr/td[1]/a/@href').extract()
            for url in links:
                yield response.follow(url = url, callback=self.parse_pages)

        def parse_pages(self, response):
            dict_temp['fighter_name'].append(response.xpath('/html/body/section/div/h2/span[1]/text()').extract()[0].strip())
            dict_temp['record'].append(response.xpath('/html/body/section/div/h2/span[2]/text()').extract()[0].strip())
            dict_temp['height'].append(response.xpath('/html/body/section/div/div/div[1]/ul/li[1]/text()').extract()[1].strip())
            dict_temp['weight'].append(response.xpath('/html/body/section/div/div/div[1]/ul/li[2]/text()').extract()[1].strip())
            dict_temp['reach'].append(response.xpath('/html/body/section/div/div/div[1]/ul/li[3]/text()').extract()[1].strip())
            dict_temp['stance'].append(response.xpath('/html/body/section/div/div/div[1]/ul/li[4]/text()').extract()[1].strip())
            dict_temp['dob'].append(response.xpath('/html/body/section/div/div/div[1]/ul/li[5]/text()').extract()[1].strip())
            dict_temp['strikes_landed_per_min'].append(response.xpath('/html/body/section/div/div/div[2]/div[1]/div[1]/ul/li[1]/text()').extract()[1].strip())
            dict_temp['strike_accuracy'].append(response.xpath('/html/body/section/div/div/div[2]/div[1]/div[1]/ul/li[2]/text()').extract()[1].strip())
            dict_temp['strikes_absorbed_per_min'].append(response.xpath('/html/body/section/div/div/div[2]/div[1]/div[1]/ul/li[3]/text()').extract()[1].strip())
            dict_temp['strike_defense'].append(response.xpath('/html/body/section/div/div/div[2]/div[1]/div[1]/ul/li[4]/text()').extract()[1].strip())
            dict_temp['takedowns_per_15_min'].append(response.xpath('/html/body/section/div/div/div[2]/div[1]/div[2]/ul/li[2]/text()').extract()[1].strip())
            dict_temp['takedown_accuracy'].append(response.xpath('/html/body/section/div/div/div[2]/div[1]/div[2]/ul/li[3]/text()').extract()[1].strip())
            dict_temp['takedown_defense'].append(response.xpath('/html/body/section/div/div/div[2]/div[1]/div[2]/ul/li[4]/text()').extract()[1].strip())
            dict_temp['submission_attempts_per_15_min'].append(response.xpath('/html/body/section/div/div/div[2]/div[1]/div[2]/ul/li[5]/text()').extract()[1].strip())

    #Define the dictionary
    dict_temp = {'fighter_name': [],
                'record':[],
                'height':[],
                 'weight':[],
                 'reach':[],
                 'stance':[],
                 'dob':[],
                 'strikes_landed_per_min':[],
                 'strike_accuracy':[],
                 'strikes_absorbed_per_min':[],
                 'strike_defense':[],
                 'takedowns_per_15_min':[],
                 'takedown_accuracy':[],
                 'takedown_defense':[],
                 'submission_attempts_per_15_min':[]
    }

    # Run the Spider
    process = CrawlerProcess()
    process.crawl(ufc_fighter_scraper)
    process.start()

    df = pd.DataFrame(dict_temp)

    return df
```

```python
df = scrape_fighter_data()
```
