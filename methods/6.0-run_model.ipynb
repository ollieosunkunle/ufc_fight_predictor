## Predict on the test dataset

```python
from pycaret.classification import *
```

```python
catboost = load_model('Final catboost Model 10Sep2020')
```

```python
predictions = predict_model(catboost, test_df)[['fighter_1','fighter_2','Label','Score']]
```

```python
predictions['Certainty_pct'] = predictions.Score.apply(lambda x: (1-x)*200 if x >= 0.5 else x*200)
```

```python
predictions['Winner'] = np.where(predictions['Label']==1,predictions['fighter_1'],predictions['fighter_2'])
```

```python
predictions['Odds'] = np.where(predictions['Label']==1,predictions['Score']/(1-predictions['Score'])*100,(1-predictions['Score'])/predictions['Score']*100)
```
