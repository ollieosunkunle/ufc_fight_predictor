```python
print('column |','win |','loss |')
print('-'*20)
for i in cols_to_plot:
    print(i,'|', int(min_long_df[(min_long_df.variable==i) & (min_long_df.results=='win')].mean()*1000)/1000,'|',
         int(min_long_df[(min_long_df.variable==i) & (min_long_df.results=='loss')].mean()*1000)/1000)
```
